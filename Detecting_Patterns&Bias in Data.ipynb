```python
date = pd.read_csv('Cleaned_GenZ_DatingApp_Data.csv')
```

```python
# Convert 'Daily_Usage_Time' to numerical values
date['Daily_Usage_Time_Hours'] = date['Daily_Usage_Time'].replace({
    '30 minutes': 0.5, '1 hour': 1, '1.5 hours': 1.5, '2 hours': 2, '3 hours': 3
})

# Histogram: Age distribution across gender
plt.figure(figsize=(10, 6))
sns.histplot(data=date, x='Age', hue='Gender', multiple='stack', bins=10)
plt.title('Age Distribution Across Genders')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Density Plot: Daily usage time across locations
plt.figure(figsize=(10, 6))
sns.kdeplot(data=date, x='Daily_Usage_Time_Hours', hue='Location', fill=True)
plt.title('Daily App Usage Time Across Locations')
plt.xlabel('Daily Usage Time (Hours)')
plt.ylabel('Density')
plt.show()
```

```python
demographic_counts = {
    'Gender': date['Gender'].value_counts(),
    'Location': date['Location'].value_counts(),
    'Age Group': pd.cut(date['Age'], [18, 22, 26, 30, 35, 40]).value_counts()
}
```
